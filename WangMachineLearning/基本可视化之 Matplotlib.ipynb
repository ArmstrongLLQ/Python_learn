```python
import pandas as pd

# index_col = 0 是说把第一列 Date 当成行标签 (index)
# parse_dates = True 是说把行标签转成 date 对象
# dayFirst = True 是说日期是 DD/MM/YYYY 这样的格式
data = pd.read_csv('S&P500.csv', index_col=0, parse_dates=True, dayfirst=True)
data.head()
```

```python
spx = data.loc['2007-01-01':'2010-01-01', ['Adj Close']]
spx.head()
```

```python
plt.plot(spx.values)
```

```python

data = pd.read_csv( 'VIX.csv', index_col=0, 
                               parse_dates=True,
                               dayfirst=True )
vix = data[['Adj Close']].loc['2007-01-01':'2010-01-01']
vix.head(3).append(vix.tail(3))
```

```python
fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(1, 1, 1)

x = spx.index
y1 = spx.values
y2 = vix.values

ax1.plot(y1, color='r',linewidth=2, linestyle='-', label='S&P500')

ax1.legend(loc='upper left', frameon=True)

ax1.set_xlim(-1, len(x)+1)

x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]

ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label, rotation=90)

ax2 = ax1.twinx()
ax2.plot(y2, linewidth=2, linestyle='-', label='VIX')
ax2.legend(loc='upper right', frameon=True)

fig.show()
```
