```python
import scikitplot as skplt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
```

```python
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train = X_train.reshape((60000, 784))
X_test = X_test.reshape((10000, 784))
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state = 1)
pca.fit(X_train)
```

```python
skplt.decomposition.plot_pca_component_variance(pca, target_explained_variance=0.75, figsize=(8, 8))

plt.show()
```

```python
skplt.decomposition.plot_pca_2d_projection(pca, X_train, y_train, figsize=(8, 8))
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(penalty='l1', solver='saga', multi_class='multinomial', n_jobs=-1)
LR.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = LR.predict(X_test)

print('Train accuracy = ', accuracy_score(y_train, LR.predict(X_train)))
print('Test accuracy = ', accuracy_score(y_test, y_pred))
```

```python
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=True, figsize=(8, 8))
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
breast_cancer.keys()
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer['data'], breast_cancer['target'], test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators=5, random_state=1, n_jobs=-1)
```

```python
skplt.estimators.plot_learning_curve(RF, X_train, y_train, cv=10, figsize=(8, 8))
plt.show()
```
