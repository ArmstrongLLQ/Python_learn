```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
np.random.seed(1031)
```

```python
iris_data = pd.read_csv('iris-data.csv', na_values=['NA'])
```

```python
iris_data.head(10)
```

```python
iris_data.describe()
```

```python
sns.pairplot(iris_data.dropna(), hue='class')  
```

```python
iris_data.loc[iris_data['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris_data.loc[iris_data['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'

iris_data['class'].unique()
```

```python
sns.pairplot(iris_data.dropna(), hue='class')  
```

```python
cond = (iris_data['class'] == 'Iris-setosa') & (iris_data['sepal_width_cm'] < 2.5)
iris_data.loc[cond]
```

```python
iris_data = iris_data.loc[~cond]
iris_data.loc[iris_data['class'] == 'Iris-setosa',
             'sepal_width_cm'].hist()
```

```python
cond = (iris_data['class'] == 'Iris-versicolor') & (iris_data['sepal_length_cm'] < 1.0)
iris_data.loc[cond]
```

```python

iris_data.loc[cond, 'sepal_length_cm'] *= 100.0

iris_data.loc[iris_data['class'] == 'Iris-versicolor', 
             'sepal_length_cm'].hist()
```
