```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x), '-', label='sinx')
ax.plot(x, np.cos(x), '--', label='cosx')
ax.legend()
ax.set(xlabel='value', ylabel='count')
ax.axis('tight')
```
